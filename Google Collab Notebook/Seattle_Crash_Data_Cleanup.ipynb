```python
# read csv file into dataframe
csv_path = '/content/drive/MyDrive/UCD Data Analytics - Project 4/SDOT_Collisions_-_All_Years.csv'
df = pd.read_csv(csv_path)
```

```python
# change column names to string format
df = pd.read_csv(csv_path, dtype={'column_name': str})
```

```python
# look at column names
df.columns
```

```python
# rename column names to be more visually appealing and clear
df = df.rename(columns={'OBJECTID': 'Object_Id', 
                        'INCKEY': 'Incident_Key', 
                        'ADDRTYPE': 'Address_Type', 
                        'COLLISIONTYPE': 'Collision_Type',
                        'INCDATE': 'Incident_Date', 
                        'FATALITIES': 'Fatalities', 
                        'INJURIES': 'Injuries', 
                        'JUNCTIONTYPE': 'Junction_Type', 
                        'LIGHTCOND': 'Light_Conditions', 
                        'LOCATION': 'Location', 
                        'PEDCOUNT': 'Pedestrian_Count', 
                        'PEDCYLCOUNT': 'Cyclist_Count',
                        'PERSONCOUNT': 'Person_Count', 
                        'ROADCOND': 'Road_Condition', 
                        'SERIOUSINJURIES': 'Serious_Injuries', 
                        'SEVERITYDESC': 'Severity_Description', 
                        'SPEEDING': 'Speeding', 
                        'UNDERINFL': 'Under_The_Influence', 
                        'VEHCOUNT': 'Vehicle_Count', 
                        'WEATHER': 'Weather',
                        'HITPARKEDCAR': 'Hit_Parked_Car',
                        'SDOT_COLDESC': 'Collision_Description'})
```

```python
# drop unnecessary columns
columns_to_drop = ['COLDETKEY', 'EXCEPTRSNDESC', 'INATTENTIONIND', 'ADDDTTM', 'DIAGRAMLINK', 'DISTANCE', 'EXCEPTRSNCODE', 'INCDTTM',
                   'INTKEY', 'PEDROWNOTGRNT', 'REPORTLINK', 'REPORTNO', 'SDOT_COLCODE', 'SDOT_COLDESC', 'SDOTCOLNUM', 'SEGKEY',
                   'SEVERITYCODE', 'ST_COLCODE', 'ST_COLDESC', 'STATUS', 'WIDTH', 'SE_ANNO_CAD_DATA', 'SEGLANEKEY', 'CROSSWALKKEY', 'SPDCASENO', 'SOURCE',
                   'SOURCEDESC', 'ADDBY', 'MODBY', 'MODDTTM', 'Speeding']

df = df.drop(columns=columns_to_drop, errors='ignore')
```

```python
# look at remaining column names
df.columns
```

```python
# view complete dataframe 
df
```

```python
# create copy of dataframe for model training
model_df = df.copy()
model_df.head()
```

```python
# look at value counts for our target variable
model_df['Severity_Description'].value_counts()
```

```python
# change values to numeric
model_df['Severity_Description'] = model_df['Severity_Description'].replace(['Property Damage Only Collision', 'Injury Collision', 'Serious Injury Collision', 'Fatality Collision'], [0,1,2,3])
model_df['Severity_Description'] = model_df['Severity_Description'].replace('Unknown', 1)
model_df
```

```python
# confirm values for target variable
model_df['Severity_Description'].value_counts()
```

```python
# view datatypes for dataframe
model_df.dtypes
```

```python
# drop columns unnecessary for training
model_df = model_df.drop(['Longitude', 'Latitude', 'Object_Id', 'Incident_Key', 'Incident_Date', 'Location', 'Fatalities', 'Injuries', 'Serious_Injuries'], axis=1)
model_df.head()
```

```python
# change yes or no columns to a single numeric column
model_df['Hit_Parked_Car'] = model_df['Hit_Parked_Car'].replace(['N', 'Y'], [0, 1])
model_df['Under_The_Influence'] = model_df['Under_The_Influence'].replace(['N', 'Y'], [0, 1])
model_df.head()
```

```python
# confirm values for yes or no columns
model_df['Hit_Parked_Car'].value_counts()
```

```python
# look at remaining column datatypes
model_df.dtypes
```

```python
# convert remaining categorical columns to numeric with get_dummies
converted_df = pd.get_dummies(model_df)
converted_df.head()
```

```python
# declare target variable and features
y = converted_df['Severity_Description'].ravel()
X = converted_df.drop(columns='Severity_Description')
```

```python
# split values for trainging and testing
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# scale values with StandardScaler
scaler = StandardScaler()
X_scaler = scaler.fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# create classification model with XGBoost
xg_model = XGBClassifier(n_estimators=250, max_depth=4, learning_rate=1, objective='multi:softmax')
```

```python
# fit the data to the model
xg_model.fit(X_train_scaled, y_train)
```

```python
# make predictions from test data
xg_preds = xg_model.predict(X_test_scaled)
```

```python
# generate confusion matrix and accuracy score
cm_xg = confusion_matrix(y_test, xg_preds)
cm_xg_df = pd.DataFrame(
    cm_xg, index=["Actual 0", "Actual 1", "Actual 2", "Actual 3"], columns=["Predicted 0", "Predicted 1", "Predicted 2", "Predicted 3"]
)

acc_score_xg = accuracy_score(y_test, xg_preds)
```

```python
# display confusion matrix and accuracy score
print("Confusion Matrix")
display(cm_xg_df)
print(f"Accuracy Score : {acc_score_xg}")
print("Classification Report")
print(classification_report(y_test, xg_preds))
```

```python
# create classification model with Decision Tree
tree_model = tree.DecisionTreeClassifier()
```

```python
# fit the data to the model
tree_model = tree_model.fit(X_train_scaled, y_train)
```

```python
# make predictions from test data
tree_preds = tree_model.predict(X_test_scaled)
```

```python
# generate confusion matrix and accuracy score
cm_tree = confusion_matrix(y_test, tree_preds)
cm_tree_df = pd.DataFrame(
    cm_tree, index=["Actual 0", "Actual 1", "Actual 2", "Actual 3"], columns=["Predicted 0", "Predicted 1", "Predicted 2", "Predicted 3"]
)

acc_score_tree = accuracy_score(y_test, tree_preds)
```

```python
# display confusion matrix and accuracy score
print("Confusion Matrix")
display(cm_tree_df)
print(f"Accuracy Score : {acc_score_tree}")
print("Classification Report")
print(classification_report(y_test, tree_preds))
```

```python
# create classification model with Random Forest
rf_model = RandomForestClassifier(n_estimators=200, random_state=17)
```

```python
# fit the data to the model
rf_model = rf_model.fit(X_train_scaled, y_train)
```

```python
# make predictions from test data
predictions = rf_model.predict(X_test_scaled)
```

```python
# generate confusion matrix and accuracy score
cm = confusion_matrix(y_test, predictions)
cm_df = pd.DataFrame(
    cm, index=["Actual 0", "Actual 1", "Actual 2", "Actual 3"], columns=["Predicted 0", "Predicted 1", "Predicted 2", "Predicted 3"]
)

acc_score = accuracy_score(y_test, predictions)
```

```python
# display confusion matrix and accuracy score
print("Confusion Matrix")
display(cm_df)
print(f"Accuracy Score : {acc_score}")
print("Classification Report")
print(classification_report(y_test, predictions))
```

```python
# create function to convert numeric prediction back to string value
def get_severity(number):
  severities = {
      0: "Property Damage Only Collision",
      1: "Injury Collision",
      2: "Serious Injury Collision",
      3: "Fatality Collision"
  }
  return severities.get(number[0])
```
